## Top 5 Countries by Immigration in 2024

This analysis identifies the five European countries that received the highest number of immigrants in 2024.

The data is filtered to include only records from 2024, sorted in descending order by the number of immigrants, and the top five countries are displayed.

```python
# Filter data for 2024

df_2024 = df[df["Year"] == 2024]

# Display the five countries with the highest immigration

df_2024.sort_values("Immigrants", ascending=False).head(5)
```

```python
immigration_by_year = (
    df.groupby("Year")["Immigrants"]
      .sum()
      .reset_index()
)
```

# Question 1: How has immigration evolved in Europe between 2015 and 2024?

```python
plt.figure(figsize=(10, 5))

plt.plot(
    immigration_by_year["Year"],
    immigration_by_year["Immigrants"],
    marker="o"
)

plt.title("Immigration Trend in Europe (2015–2024)")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")

plt.grid(True)

plt.show()
```

### Interpretation

Immigration across the selected European countries remained relatively stable between 2015 and 2019, followed by a noticeable decline in 2020. From 2021 onwards, immigration increased sharply, reaching its highest level in 2022 before decreasing slightly in 2023 and 2024.

Despite the decline after 2022, immigration levels remained well above those observed before 2021.

### Conclusion

Overall, immigration in Europe increased considerably after 2020, reaching a peak in 2022. Although immigration declined slightly in the following two years, it remained well above pre-2021 levels by the end of the study period.

# Question 2

## Which countries received the highest number of immigrants in 2024?

```python
plt.figure(figsize=(12, 6))

plt.barh(
    top_10_immigration_2024["Country"],
    top_10_immigration_2024["Immigrants"]
)

plt.gca().invert_yaxis()

plt.title("Top 10 Countries by Immigration (2024)")
plt.xlabel("Number of Immigrants")
plt.ylabel("Country")

plt.show()
```

### Interpretation

The distribution of immigrants across European countries in 2024 was highly concentrated. Spain and Germany clearly stood out as the main destinations, receiving significantly more immigrants than any other country in the dataset. France ranked third, followed by Poland and the Netherlands, while the remaining countries in the top 10 recorded considerably lower immigration figures.

Overall, the chart shows a substantial gap between the first two countries and the rest, indicating that immigration flows in 2024 were concentrated in a small number of destinations.

    ### Conclusion
    
    Spain and Germany received the highest number of immigrants in 2024, with more than one million and 950,000 immigrants, respectively. Both countries recorded substantially higher immigration figures than the rest of the top 10, while France ranked third with just over 350,000 immigrants.

# Project Question 3

## Which countries experienced the highest immigration growth between 2015 and 2024?

```python
immigration_pivot = df.pivot(
    index="Country",
    columns="Year",
    values="Immigrants"
)
```

```python
immigration_pivot["growth_(%)"] = (
    (immigration_pivot[2024] - immigration_pivot[2015])
    / immigration_pivot[2015]
) * 100
```

```python
immigration_pivot = immigration_pivot.sort_values(
    "growth_(%)",
    ascending=False
).head(10)
```

```python
plt.figure(figsize=(12, 6))

plt.barh(
    immigration_pivot.index,
    immigration_pivot["growth_(%)"]
)

plt.gca().invert_yaxis()

plt.title("Top 10 Countries by Immigration Growth (2015–2024)")
plt.xlabel("Immigration Growth (%)")
plt.ylabel("Country")

plt.show()
```

### Interpretation

The chart compares the percentage growth in immigration across European countries between 2015 and 2024. The results reveal considerable differences, with some countries recording much faster growth than others over the study period.

It is important to note that the ranking is based on percentage growth rather than the total number of immigrants. As a result, countries with relatively low immigration levels in 2015 could rank highly if they experienced substantial increases by 2024.

### Conclusion

Immigration growth between 2015 and 2024 was not evenly distributed across Europe. While some countries experienced remarkable percentage increases, these were not necessarily the countries receiving the largest numbers of immigrants, highlighting the difference between relative growth and absolute immigration levels.

# Project Question 4

## Which countries experienced the largest absolute increase in immigration between 2015 and 2024?

```python
absolute_growth_pivot = df.pivot(
    index="Country",
    columns="Year",
    values="Immigrants"
)
```

```python
absolute_growth_pivot["absolute_increase"] = absolute_growth_pivot[2024] - absolute_growth_pivot[2015]
```

```python
absolute_growth_pivot=absolute_growth_pivot.sort_values("absolute_increase", ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))

plt.barh(
    absolute_growth_pivot.index,
    absolute_growth_pivot["absolute_increase"]
)

plt.gca().invert_yaxis()

plt.title("Top 10 Countries by Absolute Increase in Immigration (2015–2024)")
plt.xlabel("Absolute Increase in Immigrants")
plt.ylabel("Country")

plt.show()
```

### Interpretation

The chart compares the absolute increase in the number of immigrants between 2015 and 2024 across European countries. Unlike the previous analysis based on percentage growth, this comparison highlights the countries that added the largest number of immigrants over the study period.

Countries with larger populations and higher immigration volumes tend to rank higher in absolute terms, even if their percentage growth was more moderate.

### Conclusion

The largest absolute increases in immigration were concentrated in a small number of European countries. These results highlight the main destinations where immigration expanded the most in numerical terms, complementing the percentage growth analysis presented in the previous question.

# Project Question 5

## How did the COVID-19 pandemic affect immigration in Europe?

```python
plt.figure(figsize=(10, 5))

plt.plot(
    immigration_by_year["Year"],
    immigration_by_year["Immigrants"],
    marker="o"
)

plt.axvline(

    x=2020,

    color="red",

    linestyle="--",

    linewidth=1,

    label="COVID-19"

)

plt.title("Impact of the COVID-19 Pandemic on Immigration in Europe")
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")

plt.grid(True)

plt.legend()

plt.show()
```

### Interpretation

The chart illustrates a clear change in immigration patterns during the COVID-19 pandemic. Immigration declined noticeably in 2020, coinciding with travel restrictions and border closures implemented across Europe.

From 2021 onwards, immigration recovered rapidly, reaching its highest level in 2022 before stabilizing at levels that remained significantly higher than those observed before the pandemic.

### Conclusion

The COVID-19 pandemic had a clear short-term impact on immigration in Europe, causing a temporary decline in 2020. However, this effect was followed by a strong recovery, with immigration exceeding pre-pandemic levels by the end of the study period.

# Project Question 6

## Is there a relationship between immigration and homicide rates?

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    df["Immigrants"],
    df["Homicide"]
)

plt.title("Immigration vs Homicide Rate in Europe (2015–2024)")
plt.xlabel("Number of Immigrants")
plt.ylabel("Homicide Rate (per 100,000 inhabitants)")

plt.grid(True)

plt.show()
```

### Interpretation

The scatter plot shows a high level of dispersion and no clear linear pattern between immigration and homicide rates across the observations. Although some countries received significantly higher numbers of immigrants than others, higher immigration does not appear to be consistently associated with higher homicide rates.

### Conclusion

Based on the visual inspection of the data, there is no clear evidence of a strong relationship between immigration and homicide rates in Europe during the period 2015–2024. The observed pattern suggests that immigration alone does not explain differences in homicide rates across the dataset.

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    df["Immigrants"],
    df["Theft"]
)

plt.title("Immigration vs Theft Rate in Europe (2015–2024)")
plt.xlabel("Number of Immigrants")
plt.ylabel("Theft Rate (per 100,000 inhabitants)")

plt.grid(True)

plt.show()
```

### Interpretation

The scatter plot shows a high level of dispersion and no clear linear pattern between immigration and theft rates across the observations. Although theft rates vary considerably between countries and years, higher immigration does not appear to be consistently associated with higher theft rates.

### Conclusion

Based on the visual inspection of the data, there is no clear evidence of a strong relationship between immigration and theft rates in Europe during the period 2015–2024. The observed distribution suggests that factors other than immigration are likely to play a more significant role in explaining differences in theft rates across the dataset.

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    df["Immigrants"],
    df["Rape"]
)

plt.title("Immigration vs Rape Rate in Europe (2015–2024)")
plt.xlabel("Number of Immigrants")
plt.ylabel("Rape Rate (per 100,000 inhabitants)")

plt.grid(True)

plt.show()
```

### Conclusion

Based on the visual inspection of the data, there is no clear evidence of a strong relationship between immigration and rape rates in Europe during the period 2015–2024. The observed variation indicates that immigration alone does not explain differences in rape rates across the dataset.

# Final Conclusions

This project explored the relationship between immigration and crime across 28 European countries between 2015 and 2024 using SQL, Python and Tableau.

The analysis showed that immigration increased substantially during the study period, particularly after the COVID-19 pandemic, with Spain, Germany and France receiving the largest numbers of immigrants in 2024. At the same time, crime indicators showed different patterns across countries, with no consistent evolution comparable to immigration trends.

The visual analyses did not reveal a clear relationship between immigration and homicide, theft or rape rates. Scatter plots showed a high level of dispersion, suggesting that countries with higher immigration levels do not systematically experience higher crime rates. The comparison of selected case studies reinforced this finding, as the countries with the highest immigration did not coincide with those reporting the highest crime rates.

Overall, the results suggest that immigration alone is not sufficient to explain differences in crime rates across European countries. Crime is a complex phenomenon influenced by multiple social, economic and demographic factors. Therefore, the findings of this exploratory analysis should be interpreted with caution and should not be considered evidence of a causal relationship.

This project demonstrates how data analysis techniques can be used to explore complex social issues while highlighting the importance of interpreting statistical results objectively and acknowledging the limitations of the available data.

## Limitations

- The analysis is exploratory and does not establish causal relationships.

- The study uses aggregated country-level data, which may hide regional differences within each country.
